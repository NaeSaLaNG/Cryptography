Задание: Реализация шифра Цезаря (или другого исторического шифра)

Цель:
Разработать программу для шифрования и дешифрования текстовых сообщений с использованием одного из исторических шифров, например, шифра Цезаря.

Описание задания:
Шифр Цезаря — это простой метод шифрования, где каждая буква исходного текста сдвигается на определённое количество позиций в алфавите. В этом задании студенты должны реализовать алгоритм шифрования и дешифрования с возможностью выбора шага сдвига.

Этапы выполнения задания:
Выбор шифра:

Выберите один из следующих исторических шифров для реализации:
Шифр Цезаря (сдвиг каждой буквы на фиксированное количество позиций).
Шифр Атбаш (инверсия алфавита).
Шифр Виженера (использование ключевого слова для шифрования).
Опишите кратко алгоритм работы выбранного шифра в комментариях к коду.
Реализация программы:

Реализуйте функцию шифрования, которая принимает на вход текст и ключ (для шифра Цезаря это будет число, а для других шифров — ключевое слово).
Реализуйте функцию дешифрования, которая по зашифрованному тексту и ключу восстанавливает исходное сообщение.
Требования к программе:

Программа должна корректно обрабатывать как прописные, так и строчные буквы.
Необходимо игнорировать символы, не являющиеся буквами (например, пробелы, цифры, знаки препинания).
Реализуйте возможность выбора языка шифрования (например, русский или английский алфавиты).
Тестирование программы:

Напишите тесты для проверки работы программы с разными значениями ключа.
Проверьте работу программы на нескольких примерах, в том числе на коротких и длинных сообщениях.

```python
def caesar_cipher_encrypt(text, shift):
    alphabet_upper_ru = 'АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ'
    alphabet_lower_ru = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'
    alphabet_upper_en = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    alphabet_lower_en = 'abcdefghijklmnopqrstuvwxyz'
    
    encrypted_text = []

    for char in text:
        # Выбор языка
        if char in alphabet_lower_en + alphabet_upper_en:
            alphabet_upper = alphabet_upper_en
            alphabet_lower = alphabet_lower_en
        elif char in alphabet_lower_ru + alphabet_upper_ru:
            alphabet_upper = alphabet_upper_ru
            alphabet_lower = alphabet_lower_ru
            
        if char.isupper() and char in alphabet_upper:
            # Сдвиг для прописных букв
            new_pos = (alphabet_upper.index(char) + shift) % len(alphabet_upper)
            encrypted_text.append(alphabet_upper[new_pos])
        elif char.islower() and char in alphabet_lower:
            # Сдвиг для строчных букв
            new_pos = (alphabet_lower.index(char) + shift) % len(alphabet_lower)
            encrypted_text.append(alphabet_lower[new_pos])
        else:
            # Оставляем символы, которые не являются буквами, без изменений
            encrypted_text.append(char)

    return ''.join(encrypted_text)

def caesar_cipher_decrypt(text, shift):
    alphabet_upper_ru = 'АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ'
    alphabet_lower_ru = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'
    alphabet_upper_en = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    alphabet_lower_en = 'abcdefghijklmnopqrstuvwxyz'
    
    decrypted_text = []

    for char in text:
        # Выбор языка
        if char in alphabet_lower_en + alphabet_upper_en:
            alphabet_upper = alphabet_upper_en
            alphabet_lower = alphabet_lower_en
        elif char in alphabet_lower_ru + alphabet_upper_ru:
            alphabet_upper = alphabet_upper_ru
            alphabet_lower = alphabet_lower_ru
        
        if char.isupper() and char in alphabet_upper:
            # Обратный сдвиг для прописных букв
            new_pos = (alphabet_upper.index(char) - shift) % len(alphabet_upper)
            decrypted_text.append(alphabet_upper[new_pos])
        elif char.islower() and char in alphabet_lower:
            # Обратный сдвиг для строчных букв
            new_pos = (alphabet_lower.index(char) - shift) % len(alphabet_lower)
            decrypted_text.append(alphabet_lower[new_pos])
        else:
            # Оставляем символы, которые не являются буквами, без изменений
            decrypted_text.append(char)

    return ''.join(decrypted_text)

text = "Привет, Мир!, Fdsfasf"
shift = 3

encrypted = caesar_cipher_encrypt(text, shift)
decrypted = caesar_cipher_decrypt(encrypted, shift)

print(f"Оригинальный текст: {text}")
print(f"Зашифрованный текст: {encrypted}")
print(f"Расшифрованный текст: {decrypted}")
```
